```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
import pandas
```

```python
MAIN = "../src/main"
TEST_DIR = "../tests/"

def run_test(test_name: str, reps: int): 
    !{MAIN} {TEST_DIR + test_name} {reps}

def read_result(test_result: str):
    return pandas.read_csv(TEST_DIR + test_result, names=["ranking"])

def absolute_error(test_result: str, expected_result: str):
    test_table = read_result(test_result)
    expected_table = read_result(expected_result)
    return abs(test_table - expected_table)
```
