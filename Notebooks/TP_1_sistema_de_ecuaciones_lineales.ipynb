## Error absoluto

Comparamos error absoluto de nuestra implementacion con los resultados de la catedra.

```python
def compare(test_name: str):
    error = absolute_error(test_name + "_salida.out", test_name + ".out")
    df = pandas.DataFrame()
    df["Error_" + test_name] = error["ranking"]
    return df
```

```python
a = compare("test_15_segundos.txt")
b = compare("test_30_segundos.txt")
c = compare("test_aleatorio_desordenado.txt")
d = compare("test_completo.txt")
e = compare("test_sin_links.txt")
f = compare("test_trivial.txt")
df_total = pandas.concat([b, a, c, d, e, f], axis = 1)
df = pandas.DataFrame(df_total[1:].mean())
df.style.background_gradient()
```

```python
def read_result_tiempos(test_result: str):
    return pd.read_csv(TEST_DIR + test_result)
```

```python
tiempo_15 = float(np.mean(read_result_tiempos("test_15_segundos.txt_tiempo.out"))[0]/1000)
tiempo_30 = float(np.mean(read_result_tiempos("test_30_segundos.txt_tiempo.out"))[0]/1000)
tiempo_lil_15 = float(np.mean(read_result_tiempos("test_15_segundos.txt_tiempo_lil.out"))[0]/1000)
tiempo_lil_30 = float(np.mean(read_result_tiempos("test_30_segundos.txt_tiempo_lil.out"))[0]/1000)

tiempo_lil_30
```

```python
labels = ['test_15_segundos', 'test_30_segundos']
men_means = [tiempo_15, tiempo_30]
women_means = [tiempo_lil_15, tiempo_lil_30]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Matriz convencional')
rects2 = ax.bar(x + width/2, women_means, width, label='Matriz LIL con listas enlazadas')

ax.set_ylabel('Tiempo (s.)')
ax.set_title('Tiempos')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

# normas

```python
def compare_norms(test_name: str):
    error = absolute_error(test_name + "_salida_error.out", test_name + ".out")
    df = pandas.DataFrame()
    df["Error_" + test_name] = error["error"]
    return df
```

```python
a = compare_norms("test_15_segundos.txt")
b = compare_norms("test_30_segundos.txt")
c = compare_norms("test_aleatorio_desordenado.txt")
d = compare_norms("test_completo.txt")
e = compare_norms("test_sin_links.txt")
f = compare_norms("test_trivial.txt")
df_total = pandas.concat([b, a, c, d, e, f], axis = 1)
df = pandas.DataFrame(df_total[1:].mean())
df.style.background_gradient()
```
